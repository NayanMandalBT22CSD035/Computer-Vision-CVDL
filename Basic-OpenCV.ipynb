```python
img = cv2.imread('dog.jpg')
```

```python
# Convert from BGR to RGB for proper color display
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img_rgb)
plt.show()
```

```python
# Apply Gaussian Blur
gaussian_blur = cv2.GaussianBlur(img_rgb, (15, 15), 0)
```

```python
plt.imshow(gaussian_blur)
plt.show()
```

```python
# Apply Gaussian Blur
gaussian_blur = cv2.GaussianBlur(img_rgb, (3, 3), 0)
plt.imshow(gaussian_blur)
plt.show()
```

```python
# Apply Sharpening Filter
kernel_sharpening = np.array([[0, -1, 0],
                              [-1, 5, -1],
                              [0, -1, 0]])
sharpened = cv2.filter2D(img_rgb, -1, kernel_sharpening)
```

```python
plt.imshow(sharpened)
plt.show()
```

```python
# Create a Contrast Filter
alpha = 1.5  # Contrast control (1.0-3.0) - increase for higher contrast
beta = 0     # Brightness control (0-100) - adjust if needed

# Apply Contrast Filter
contrast_enhanced = cv2.convertScaleAbs(img_rgb, alpha=alpha, beta=beta)
```

```python
plt.imshow(contrast_enhanced)
plt.show()
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray,cmap='gray')
plt.show()
```

```python
# Apply Histogram Equalization
equalized_image = cv2.equalizeHist(gray)
```

```python
plt.imshow(equalized_image,cmap='gray')
plt.show()
```

```python
# Apply a Box Filter
kernel_size = (5, 5)  # Define the kernel size (width, height)
box_filtered = cv2.blur(img_rgb, kernel_size)
```

```python
plt.imshow(box_filtered)
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image in grayscale
image_path = "your_image.jpg"  # Replace with the path to your image
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

if image is None:
    raise FileNotFoundError("Image not found. Please check the path.")

# Apply Sobel filters to calculate gradients in the x and y directions
sobel_x = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=3)  # Gradient in x direction
sobel_y = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=3)  # Gradient in y direction

# Calculate the gradient magnitude
gradient_magnitude = np.sqrt(sobel_x**2 + sobel_y**2)
# Normalize to 0-255
gradient_magnitude = np.uint8(255 * gradient_magnitude / np.max(gradient_magnitude))

# Display the results
plt.figure(figsize=(10, 10))
plt.subplot(2, 2, 1)
plt.title("Original Image")
plt.imshow(image, cmap='gray')
plt.axis('off')

plt.subplot(2, 2, 2)
plt.title("Sobel X")
plt.imshow(np.abs(sobel_x), cmap='gray')
plt.axis('off')

plt.subplot(2, 2, 3)
plt.title("Sobel Y")
plt.imshow(np.abs(sobel_y), cmap='gray')
plt.axis('off')

plt.subplot(2, 2, 4)
plt.title("Gradient Magnitude")
plt.imshow(gradient_magnitude, cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()
```
